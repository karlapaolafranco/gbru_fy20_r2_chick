```python
%matplotlib inline
#importing programs
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#visualizing using seaborn 
#update fifth attempt file on ubuntu and then change this line of code to pd.read_csv
sns.set_theme(style="whitegrid", font_scale = 1.5)
data3 =pd.read_csv("Fifth_Attempt.csv")
data3 =data3.sort_values(by="#sum_counts", ascending=False)
data3
#data3 = data3.sort_values(by="#sum_counts_chick1", ascending=False)
```

```python
fig=plt.gcf()
fig.set_size_inches(20,20)
ax = sns.stripplot(x="#sum_counts", y = "#Datasets", hue = "#sample", dodge=True, orient= "h", data=data3)
plt.xlabel("Normalized Counts", fontsize =10)
x_ticks= [i for i in range(-1, 7)]
ax.xaxis.set_ticks(x_ticks)
fig.autofmt_xdate()
plt.xticks(fontsize=10)
plt.ylabel("Taxa", fontsize=10)
plt.yticks(fontsize=10)
plt.legend(["Chick 1","Chick 2"])
plt.savefig("fifth_attempt_stripplot.png", bbox_inches='tight', dpi=300)
```
