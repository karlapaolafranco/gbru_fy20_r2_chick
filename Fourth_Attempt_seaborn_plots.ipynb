```python
%matplotlib inline
#importing programs
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#visualizing using seaborn 
sns.set_theme(style="whitegrid", font_scale = 1.5)
data3 =pd.read_csv('Fifth_Attempt_data_MR&clr_chick1_chick2.csv')
data3
```

```python
#make PairGrid_brackets and a # tells the program which specific column to look at 
#Do not separate this three lines of data, otherwise you will get ValueError: Axes instance argument was not found in a figure

data3_1 = sns.PairGrid(data3, x_vars= data3.columns[2:], y_vars= data3.columns[1], height=20, aspect= 1.7)
data3_1.set(xlim=(-4, 4), xlabel= "", ylabel= "Species")
data3_1.map(sns.stripplot, size=10, orient="h", palette="flare_r", linewidth=20, edgecolor="w")

titles=["Chick 1", "Chick 2"]
for ax, title in zip(data3_1.axes.flat,titles):
    ax.set(title=title)
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

sns.despine(left=True, bottom=True)
plt.savefig("fifth_attempt_dot_plot_chick1_chick2.png", bbox_inches='tight', dpi=300)
```
