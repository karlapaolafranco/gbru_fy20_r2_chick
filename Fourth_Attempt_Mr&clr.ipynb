```python
import pandas as pd
```

```python
df_chick1=pd.read_clipboard()
df_chick1 
```

```python
df_chick2=pd.read_clipboard()
df_chick2
```

```python
dfmerged = df_chick1.merge(df_chick2, on = "#Datasets", how = 'outer')
dfmerged 
```

```python
#this is a panda dataframe; if you are using a numpy dataframe then you cannot use fillna, instead use X.replace(np.nan,0)
dfmerged = dfmerged.fillna(0)
dfmerged = dfmerged.T
dfmerged
```

```python
import numpy as npy
```

```python
from skbio.stats.composition import multiplicative_replacement as MR
```

```python
#MAKE AN ARRAY IN ORDER TO APPLY MR
ind=["#sum_counts_chick1", "#sum_counts_chick2"]#look for columns 
X = npy.array(dfmerged.T[ind], dtype="float64")
X = X.T
X
```

```python
MultX=MR(X)
MultX
```

```python
from skbio.stats.composition import clr
```

```python
clrX=clr(MultX)

clrX
```

```python
dfmerged = dfmerged.T
clrX = npy.transpose(clrX)
clrX
```

```python
dfmerged["#sum_counts_chick1"] = clrX[:,0]
dfmerged["#sum_counts_chick2"] = clrX[:,1]
dfmerged
```

```python
dfmerged.to_csv('Fifth_Attempt_data_MR&clr_chick1_chick2.csv')
```
