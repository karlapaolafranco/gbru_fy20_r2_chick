```python
%matplotlib inline
#importing programs
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D
```

```python
#visualizing using seaborn 
#update fifth attempt file on ubuntu and then change this line of code to pd.read_csv
sns.set_theme(style="whitegrid", font_scale = 1.5)
data3 =pd.read_clipboard()
data3
#data3 = data3.sort_values(by="#sum_counts_chick1", ascending=False)
```

```python
fig=plt.gcf()
fig.set_size_inches(20,40)
sns.set_theme(style="whitegrid", font_scale = 2)
ax = sns.stripplot(x="#sum_counts", y = "#Datasets", hue = "#sample", jitter=False, dodge= False, orient= "h",palette =("blue", "firebrick"), size=8, edgecolor="gray", data=data3)
plt.xlabel("Normalized Counts", fontsize =22)
x_ticks= [i for i in range(-1, 7)]
ax.xaxis.set_ticks(x_ticks)
fig.autofmt_xdate()
plt.xticks(fontsize= 15)
plt.ylabel("Taxa", fontsize= 22)
plt.yticks(fontsize=15)


legend_element_1 = Line2D([0],[0], color='blue', marker='o', label="Chick 1", markersize = 10)
legend_element_2 = Line2D([0],[0], color='firebrick', marker='o', label="Chick 2",markersize = 10)

plt.legend(handles =[legend_element_1, legend_element_2])                  
#blue_patch= mpatches.Patch(color = ("blue"), label =("Chick 1"), marker = 'O')
#yellow_patch= mpatches.Patch(color = ("firebrick"), label =("Chick 2"), marker = 'O')
#plt.legend(handles =[red_patch, yellow_patch])

for i in data3["#Datasets"]:
    plt.axhline(y=i, linewidth= 5, linestyle = "-", color="whitesmoke")
    
plt.savefig("fifth_attempt_stripplot_Orderedbyphylogeny.png", bbox_inches='tight', dpi=300)
```
